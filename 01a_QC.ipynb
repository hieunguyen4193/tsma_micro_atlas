```python
import pandas as pd 
import numpy as np 
import os 
import pathlib 
from tqdm import tqdm 

import re

from helper_functions import * 
path_to_all_fa = "/Volumes/HNSD01/storage/ref/hg19"
outdir = "/Volumes/HNSD01/outdir"
# outdir = "/Volumes/HNSD03/outdir"

metadata = pd.read_excel("/Users/hieunguyen/src/tsma_micro_atlas/metadata/metadata_TSMA_R8281_R8282.xlsx")

outputdir = os.path.join(outdir, "output_methyl_ReadBased_Models")
path_to_01_output = os.path.join("/Volumes/HNWD02/outdir/output_methyl_ReadBased_Models", "01_output")
path_to_qc_output = os.path.join(outputdir, "01a_output")
os.system(f"mkdir -p {path_to_qc_output}")

panel = "TSMA_panel"
all_regions = [item.name for item in pathlib.Path(path_to_01_output, panel).glob("*") if ".DS_Store" not in item.name]
for region in tqdm(all_regions):
    all_files = [item for item in pathlib.Path(os.path.join(path_to_01_output, panel, region)).glob("*tmpdf.csv*") if "._" not in item.name]
    if len(all_files) != 0:
        tmp_countdf = pd.DataFrame()
        for file in tqdm(all_files):
            tmpdf_rows = int(os.popen(f"wc -l < {file}").read().strip()) - 1  # subtract 1 for header
            sampleid = file.name.split(".tmpdf")[0]
            tmp_countdf = pd.concat([tmp_countdf,
                                    pd.DataFrame({"LABCODE": sampleid, region: tmpdf_rows}, index=[0])],
                                    ignore_index=True)
        metadata = metadata.merge(tmp_countdf, right_on = "LABCODE", left_on = "LABCODE", how="left")
        metadata = metadata.fillna(0)
```

```python
metadata.to_csv(os.path.join(path_to_qc_output, "metadata_with_region_read_counts.csv"), index=False)
```
